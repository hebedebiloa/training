```python
#importation des librairies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#on telecharge les données
customers=pd.read_csv('https://cocl.us/customer_dataset/customer_segmentation.csv')
```

```python
customers.head(10)
```

```python
#On va supprimer le champ 'address' etant donné que l'algorithme k-means ne traite que des distances donc des valeurs quantitatives
customers=customers.drop('Address', axis=1)
customers.head()
```

```python
#On va standardiser les données afin de donner le meme poids a chaque point
X=customers.values[:,1:]
print(X)
```

```python
#remplacer les NaN par 0
X=np.nan_to_num(X)
print(X)
```

```python
#On va normalisé les données (diistribution moyenne=0 et std=1) avant d'appliquer la technique de ML
from sklearn.preprocessing import StandardScaler
cluster_data=StandardScaler().fit_transform(X)
print(cluster_data)
```

```python
#Partitionnement en 3 groupes
from sklearn.cluster import KMeans
```

```python
k=3
k_means=KMeans(init='k-means++',n_clusters=k,n_init=12)
k_means.fit(cluster_data) #apploication de l'algrorithme aux données standardisées
labels=k_means.labels_ #obtention des etiquettes

print(labels)
```

```python
# On assigne chaque client a un groupe
customers['labels']=labels
customers.head()
```

```python
#on regroupe par label
grps=customers.groupby("labels")
grps.count()
```
